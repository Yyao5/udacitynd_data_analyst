## Clean Phase

```python
# clean copies
twitter_archive_clean = twitter_archive
image_prediction_clean = image_prediction
tweet_info_clean = tweet_info
```

```python
twitter_archive['name'].value_counts()
```

```python
twitter_archive.info()
```

```python
# remove retweets and confirm
twitter_archive_clean = twitter_archive_clean[pd.isna(twitter_archive_clean['retweeted_status_user_id'])]
twitter_archive_clean['retweeted_status_user_id'].value_counts()
```

```python
def change_name_to_None(df, word):
    num = len(df)
    for i in range(num):
        if df.iloc[i]['name'] == word:
            df.iloc[i]['name'] = 'None'
    return df

twitter_archive_clean = change_name_to_None(twitter_archive_clean, 'a')
twitter_archive_clean = change_name_to_None(twitter_archive_clean, "the")
twitter_archive_clean = change_name_to_None(twitter_archive_clean, "an")
        
```

```python
twitter_archive_clean['name'].value_counts()
```

```python
# complete URL column
def add_urls(row):
       
        tweet_id = row['tweet_id']
        # make id into new URL
        row['expanded_urls'] = 'https://twitter.com/dog_rates/status/{}'.format(tweet_id)
        return row

twitter_archive_clean = twitter_archive_clean.apply(add_urls, axis=1)
```

```python
# Standardize the capitalization of name 
twitter_archive_clean['name'] = twitter_archive_clean['name'].apply(lambda x: x.capitalize()) 
```

```python
def add_types(df):
    
    if 'puppers' in df['text']:
        df['pupper'] = 'pupper'
    if 'floofers' in df['text']:
        df['floofer'] = 'floofer'
    if 'puppos' in df['text']:
        df['puppo'] = 'puppo' 
    if 'doggos' in df['text']:
        df['doggo'] = 'doggo' 
        
    # return row whether or not it was updated    
    return df


# Save dataframe with missing urls
twitter_archive_clean = twitter_archive_clean.apply(add_types, axis=1)
```

```python
# convert types from wide to long format
def wide_to_long(df):
    
    if df['pupper'] == 'pupper':
        df['type'] = 'pupper'
    if df['floofer'] == 'floofer':
        df['type'] = 'floofer'
    if df['puppo'] == 'puppos':
         df['type'] = 'puppo'
        
    # return row whether or not it was updated    
    return df


# Save dataframe with missing urls
twitter_archive_clean = twitter_archive_clean.apply(wide_to_long, axis=1)
twitter_archive_clean = twitter_archive_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis=1)
```

```python
# convert object to datetime
twitter_archive_clean.timestamp = pd.to_datetime(twitter_archive_clean.timestamp)
```

```python
# Standardize the capitalization of name 
image_prediction['p1'] = image_prediction['p1'].apply(lambda x: x.capitalize()) 
image_prediction['p2'] = image_prediction['p2'].apply(lambda x: x.capitalize()) 
image_prediction['p3'] = image_prediction['p3'].apply(lambda x: x.capitalize()) 
```

```python
# merge the cleaned datasets
clean_final = pd.merge(twitter_archive_clean, image_prediction, how='left', on=['tweet_id'])
# rename id in tweet_info
tweet_info.rename(columns={'id':'tweet_id'}, inplace=True)
clean_final = pd.merge(clean_final, tweet_info, how='left', on=['tweet_id'])
```

### Most Common Breed
#### ... is GOLDEN RETRIEVER!


```python
clean_final['p1_dog'].value_counts()
```

```python
dog_type = []
confidence_list = []

# summarize the prediction of dog type
def image(df):
    if df['p1_dog'] == True:
        dog_type.append(df['p1'])
        confidence_list.append(df['p1_conf'])
    elif df['p2_dog'] == True:
        dog_type.append(df['p2'])
        confidence_list.append(df['p2_conf'])
    elif df['p3_dog'] == True:
        dog_type.append(df['p3'])
        confidence_list.append(df['p3_conf'])
    else:
        dog_type.append('Not_dog')
        confidence_list.append('Error')

#series objects having index the image_prediction_clean column.        
clean_final.apply(image, axis=1)

#create new columns
clean_final['breed'] = dog_type
clean_final['confidence_list'] = confidence_list
```
